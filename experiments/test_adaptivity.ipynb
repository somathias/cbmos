```python
import numpy as np
import scipy.integrate as scpi
from scipy.optimize import minimize
import matplotlib.pyplot as plt
import numpy.random as npr
import os

import euler_forward as ef
import force_functions as ff
import cbmos_serial as cbmos
import cell as cl

plt.style.use('seaborn-whitegrid')
params = {'legend.fontsize': 'x-large',
          'figure.figsize': (6.75, 5),
          'axes.labelsize': 'xx-large',
          'axes.titlesize': 'x-large',
          'xtick.labelsize': 'x-large',
          'ytick.labelsize': 'x-large'}
plt.rcParams.update(params)
```

```python
# Simulation parameters
s = 1.0    # rest length
tf = 37.0  # final time
rA = 1.5   # maximum interaction distance

params = {"mu": 5.70, "s": s, "rA": rA}
t_data = np.linspace(0, tf, 100)
```

```python
solver = cbmos.CBMSolver(ff.cubic, ef.solve_ivp, 1)
```

## Fixed division times

```python
cell_list = [cl.Cell(0, [0], -6.0, True, lambda t: 6 + t)]
```

```python
os.remove('step_sizes.txt')
os.remove('time_points.txt')
```

```python
history = solver.simulate(cell_list, t_data, params, {"eps": 0.01})
```

```python
dts = np.loadtxt('step_sizes.txt')
ts = np.loadtxt('time_points.txt')
plt.plot(ts, dts, '-*')
```

```python
plt.plot(ts[:len(history)], [len(clt) for clt in history[:len(ts)]])
```

## Normally distributed division times

```python
cell_list = [cl.Cell(0, [0], -6.0, True, lambda t: npr.normal(6 + t))]
```

```python
os.remove('step_sizes.txt')
os.remove('time_points.txt')

history = solver.simulate(cell_list, t_data, params, {"eps": 0.01})
```

```python
dts = np.loadtxt('step_sizes.txt')
ts = np.loadtxt('time_points.txt')
plt.plot(ts, dts, '-*')
```

```python
plt.plot(ts[:len(history)], [len(clt) for clt in history[:len(ts)]])
```

## Exponentially distributed division times

```python
cell_list = [cl.Cell(0, [0], 0.0, True, lambda t: npr.exponential(12) + t)]
```

```python
os.remove('step_sizes.txt')
os.remove('time_points.txt')

history = solver.simulate(cell_list, t_data, params, {"eps": 0.01})
```

```python
dts = np.loadtxt('step_sizes.txt')
ts = np.loadtxt('time_points.txt')
plt.plot(ts, dts, '-*')
```
